```python
from IPython.display import Image
import pymol;from pymol import cmd
## Header: General Commands ##
# delete all objects and reset pymol
cmd.do("reinitialize")
# set the background color to white
cmd.do("bg_color white")
# make the background transparent for ray trace 
cmd.do("set ray_opaque_background, 0")
# set the ray trace mode
# normal color
cmd.do("set ray_trace_mode, 0")
# normal color + black outline
#set ray_trace_mode,  1
# black outline only
#set ray_trace_mode,  2
# turn off shadows during ray trace
cmd.do("set ray_shadows, 0")
# set the mouse mode for laptop.
#cmd.do("config_mouse one_button")
# run the python program for hydrogen bonding interactions.
#cmd.do("run list_hbonds.py")
#------put the python program list_hbonds.py
#------into the directory python.
## End of Header: General Commands ##
```

```python
cmd.set('ray_opaque_background', '0')
cmd.set('ray_trace_mode', '0');cmd.set('ray_shadow','0')
cmd.set('stick_h_scale','1');cmd.set('valence','0')
cmd.do("load files/cg.pdb,CG");cmd.do("h_add not name C1*")
cmd.alter('name H01', 'type="name=H"')
cmd.alter('name H02', 'type="name=H"')
cmd.alter('name H03', 'type="name=H"')
cmd.alter('name H04', 'type="name=H"')
cmd.alter('name H05', 'type="name=H"')
cmd.do("show sphere, CG");cmd.alter('CG', 'vdw=0.35') 
cmd.do("rebuild")
```

```python
cmd.show('stick', 'CG')
cmd.color('wheat','name C2 ')
cmd.color('wheat','name C3 ')
cmd.color('wheat','name C3 ')
cmd.color('wheat','name C4 ')
cmd.color('wheat','name C5 ')
cmd.color('wheat','name C6 ')
cmd.color('wheat','name C8 ')
cmd.color('wheat','name C1* ')
cmd.color('wheat','name C2* ')
cmd.color('wheat','name C3* ')
cmd.color('wheat','name C4* ')
cmd.color('wheat','name C5* ')
cmd.color('wheat','name C6* ')
cmd.color('red','name O*')
cmd.color('red','name N*')
cmd.color('red','name H*')
```

```python
cmd.distance('WC_hbond','name O6+N3+O2','name H', '2.4')
cmd.set('sphere_quality', '10')
cmd.set('stick_quality','20');cmd.set('sphere_quality','20')
```
