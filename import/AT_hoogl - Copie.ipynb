```python
from IPython.display import Image
import pymol;from pymol import cmd
## Header: General Commands ##
# delete all objects and reset pymol
cmd.do("reinitialize")
# set the background color to white
cmd.do("bg_color white")
# make the background transparent for ray trace 
cmd.do("set ray_opaque_background, 0")
# set the ray trace mode
# normal color
cmd.do("set ray_trace_mode, 0")
# normal color + black outline
#set ray_trace_mode,  1
# black outline only
#set ray_trace_mode,  2
# turn off shadows during ray trace
cmd.do("set ray_shadows, 0")
# set the mouse mode for laptop.
#cmd.do("config_mouse one_button")
# run the python program for hydrogen bonding interactions.
#cmd.do("run list_hbonds.py")
#------put the python program list_hbonds.py
#------into the directory python.
## End of Header: General Commands ##
```

```python
cmd.set('ray_opaque_background', '0')
cmd.set('ray_trace_mode', '0');cmd.set('ray_shadow','0')
cmd.set('stick_h_scale','1');cmd.set('valence','0')
cmd.do("load files/1gqu.pdb,duplex");
cmd.do("h_add duplex and not name OP2 or name OP1")
cmd.do("distance hbond1,duplex and resn DA,duplex and resn DT or resn BRU,2.4")
cmd.do("distance hbond1, (duplex and resn DG), (duplex and resn DC), 2.4;\
remove resn HOH")
```

```python
cmd.do("create hoogsteen, resi 1 or resi 12")
cmd.do("h_add hoogsteen")
cmd.do("distance hbond2,hoogsteen and resn DA,hoogsteen and resn DT,2.2")
cmd.do("color black, hbond2")
cmd.do("color wheat, name C* and hoogsteen")
cmd.do("color gray60, name C* and duplex")
```

```python
cmd.do("hide everything;show stick, duplex;disable duplex;disable hbond1")
cmd.do("show stick,hoogsteen;color cyan, hbond*;show dashes, hbond*;\
show labels, hbond*;set label_size, 15;rotate z,90")
cmd.set_view("1.000000000,    0.000000000,    0.000000000,\
0.000000000,    1.000000000,    0.000000000,\
0.000000000,    0.000000000,    1.000000000,\
-0.000000447,    0.000000140,  -35.039432526,\
13.939747810,   25.590175629,   17.814601898,\
32.686237335,   37.392642975,  -20.000000000")
cmd.png("at.png",400,200,ray=1);Image(filename="at.png")
```
