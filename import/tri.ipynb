######  Analyse de l'équation $-x^2-2x$

$-x^2$ est l'accélération (dérivé de la vitesse)  et $2x$ est la dérivé du de la distance sur le temps. Autrefois l'antique civilisation $dao$ parlait de quinte, quarte, tierce, seconde.

```python
def f(x):
    if x<0:
        return sqrt(-x^2-2*x)
    else:
        return -sqrt(-x^2+2*x)
a=plot(f,(x,-2,2),color='white',fill=sqrt(4-x^2),
fillcolor="lightgray",alpha=0.2)
p=plot(f,(x,-2,2),color='white',fill=-sqrt(4-x^2),
fillcolor="white",alpha=0.1)
b = circle((0,0),2,color="lightgray",alpha=0.5)
show(a+p,figsize=2,axes=False,aspect_ratio=1)
```
